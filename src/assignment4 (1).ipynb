```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import r_regression
```

```python
# Import data
df = pd.read_csv('brain_tumor_dataset.csv').dropna()

# Store relevant columns as variables
# values change the data frame to the numpy array
X = df[['Tumor_Size']].values.reshape(-1, 1)
y = df[['Survival_Rate']].values.reshape(-1, 1)

# Initialize and fit a linear regression model
# Your code goes here
linearModel = LinearRegression().fit(X, y)

print('Intercept:', linearModel.intercept_[0])
```

```python
# Fit a least squares regression model
linModel = LinearRegression()
linModel.fit(X, y)
yPredicted = linModel.predict(X)

# Graph the model
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='black', alpha=0.05, s=50)
plt.plot(X, yPredicted, color='blue', linewidth=2)
plt.xlabel('Tumor Size', fontsize=14)
plt.ylabel('Survival Rate', fontsize=14)
plt.title('Tumor Size vs  Survival Rate')
plt.grid(True)
```

```python
# Graph the residuals
# Should I use a smaller sample size for this graph and above?
plt.scatter(X, y, color='black')
plt.plot(X, yPredicted, color='blue', linewidth=2)
# plt.plot function: the first parameter is X values, the seond parameter is y values
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], yPredicted[i]], color='grey', linewidth=1)
plt.xlabel('Tumor_Growth_Rate', fontsize=14)
plt.ylabel('Survival_Rate', fontsize=14)
```

```python
#reference
plt.figure(figsize=(10, 6))
plt.hexbin(X.flatten(), y.flatten(), gridsize=50, cmap='Blues')
plt.colorbar(label='Count in bin')
plt.xlabel('Tumor Size')
plt.ylabel('Survival Rate')
plt.title('Hexbin: Tumor Size vs Survival Rate')
plt.grid(True)
plt.show()
```

```python
#reference online
plt.figure(figsize=(10, 6))
plt.hist2d(X.flatten(), y.flatten(), bins=100, cmap='viridis')
plt.colorbar(label='Count in bin')
plt.xlabel('Tumor Size')
plt.ylabel('Survival Rate')
plt.title('2D Histogram: Tumor Size vs Survival Rate')
plt.grid(True)
plt.show()
```

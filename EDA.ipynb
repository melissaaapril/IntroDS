## Encode values

```python
# count non num columns
non_num_col = df.select_dtypes(include=['object']).columns

# print the number of string columns
print(len(non_num_col))

#print the non num col
non_num_col
```

```python
# Get Boolean columns to easily encode later
bool_col = []
mult_col = []
for x in non_num_col:
    #check if it has 2 unique vals
    if df[x].nunique() ==2:
        bool_col.append(x)
    else:
        mult_col.append(x)

bool_col
```

```python
mult_col
```

```python
# encode bianry columns separately
for col in bool_col:
    df[col] = df[col].map({df[col].unique()[0]:0, df[col].unique()[1]:1})
df
```

```python
# encode multi class columns in a one hot encoder
df = pd.get_dummies(df, columns = mult_col, drop_first = True)

df
```
